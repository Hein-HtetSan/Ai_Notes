```python
import numpy as np
import matplotlib.pyplot as plt
#metplotlib inline
```

```python
x = np.array([0,1,2,3,4,5])
y1 = x^2
y2 = 2*x

plt.scatter(x, y1, color="black")
plt.scatter(x, y2, color="black")
plt.plot(x, y1, color="red", label="x^2", linewidth=1)
plt.plot(x, y2, color="green", label="y-2x", linewidth = 3)
plt.xlabel('x-dependent')
plt.ylabel('y-denpendent')
plt.text(2,2, 'a is line of slope\n b is intercept',
        rotation=0,
        horizontalalignment="center",
        verticalalignment="top",
        multialignment="center")
plt.legend()
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([3,2,3,5,4])

plt.scatter(x,y, color="red")
plt.xlabel('independent variable')
plt.ylabel('targer (Dependent)')
plt.margins(0.5)
plt.show()
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
x = np.array([1,2,3,4,5]).reshape((-1,1))
y = np.array([3,2,3,5,4])
reg = linear_model.LinearRegression()
reg.fit(x,y)
y_predict = reg.predict(x)
reg.intercept_
#reg.coef_
```

```python
x = np.array([1,2,3,4,5])
y = np.array([3,2,3,5,4])

plt.scatter(x,y, color="red")
plt.xlabel('independent variable')
plt.ylabel('target (dependent)')
plt.margins(0.1,0.5)

y1=0.5*x+1.9
plt.scatter(x,y1, color="black")
plt.plot(x,y1, linewidth=1)
plt.show()
```

```python
r2_score(x,y)
mean_squared_error(y, y_predict)

import statistics
mse = statistics.variance(y)
mse
```
