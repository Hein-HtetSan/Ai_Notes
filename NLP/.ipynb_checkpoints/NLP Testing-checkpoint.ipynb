```python
# Access Web Text
from nltk.corpus import webtext
webtext.fileids()
```

```python
#Read raw web text
raw_webtext = webtext.raw('firefox.txt')
print(len(raw_webtext))
raw_webtext
```

```python
# Read words of web text
word_webtext = webtext.words('firefox.txt')
print(len(word_webtext))
word_webtext
```

## Tokenization raw webtext

```python
from nltk.tokenize import word_tokenize
token1_webtext = word_tokenize(raw_webtext)
print(len(token1_webtext))
token1_webtext
```

```python
from nltk.tokenize import WordPunctTokenizer
token2_webtext = WordPunctTokenizer().tokenize(raw_webtext)
print(len(token2_webtext))
token2_webtext
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer("[\w']+")
token3_webtext = tokenizer.tokenize(raw_webtext)
print(len(token3_webtext))
token3_webtext
```

```python
fdist4 = FreqDist(token3_webtext)
fdist4.most_common(20)
```

## Removing stopWords

```python
from nltk.corpus import stopwords
english_stops = set(stopwords.words('english'))
```

```python
filter_words = [word for word in token3_webtext if word not in english_stops]
fdist5 = FreqDist(filter_words)
fdist5.most_common(20)
```

```python
print(len(filter_words))
```

## Stemming and Lemmatization

```python
from nltk.stem import PorterStemmer
stem_words = []
for w in filter_words:
    rootStem = PorterStemmer().stem(w)
    stem_words.append(rootStem)
print(len(stem_words))
stem_words
```

```python
fdist6 = FreqDist(stem_words)
fdist6.most_common(20)
```

```python
from nltk.stem import WordNetLemmatizer
lem_words = []
for w in filter_words:
    rootWords = WordNetLemmatizer().lemmatize(w)
    lem_words.append(rootWords)
print(len(lem_words))
lem_words
```

```python
fdist7 = FreqDist(lem_words)
fdist7.most_common(20)
```

## POS Tagging

```python
tag_tokens = nltk.pos_tag(token3_webtext)
tag_tokens
```

## NER

```python
from nltk import ne_chunk
ne_chunk(tag_tokens[:10])
```

```python
print(ne_chunk(tag_tokens[:100]))
```

```python
def extract_ne(tags):
    tree = ne_chunk(tags, binary=True)
    return set(
        " ".join(i[0] for i in t)
        for t in tree
        if hasattr(t, "label") and t.label() == "NE"
    )
extract_ne(tag_tokens)
```
